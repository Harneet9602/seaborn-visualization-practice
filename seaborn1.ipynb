```python
import seaborn as sns  
import matplotlib.pyplot as plt  
import pandas as pd  
tips = sns.load_dataset('tips')  
plt.figure(figsize=(12, 6))  
scatter_plot = sns.scatterplot(  
    data=tips,  
    x='total_bill',  
    y='tip',         
    hue='day',       
    size='size',    
    sizes=(20, 200),   
    style='sex',       
    palette='deep',  
    alpha=0.7, 
)   
plt.title('Scatter Plot of Tips vs Total Bill', fontsize=16)  
plt.xlabel('Total Bill ($)', fontsize=14)  
plt.ylabel('Tip ($)', fontsize=14)  
plt.legend(title='Day and Sex', loc='upper left')  
plt.show()
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt  
import pandas as pd   
flights = sns.load_dataset('flights')  
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')  
plt.figure(figsize=(12, 6))   
line_plot = sns.lineplot(  
    data=flights_pivot,  
    dashes=False,             
    palette='deep',          
)  
plt.title('Monthly Number of Airline Passengers', fontsize=16)  
plt.xlabel('Month', fontsize=14)  
plt.ylabel('Number of Passengers', fontsize=14)  
plt.xticks(ticks=range(12), labels=flights['month'].unique(), rotation=45)   
plt.grid(True)    
plt.show()
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt  
import pandas as pd
tips = sns.load_dataset('tips')  
plt.figure(figsize=(10, 6))  
bar_plot = sns.barplot(  
    data=tips,  
    x='day',                
    y='total_bill',         
    hue='sex',              
    ci="sd",                
    palette='pastel'        
)  
plt.title('Average Total Bill by Day and Gender', fontsize=16)  
plt.xlabel('Day', fontsize=14)  
plt.ylabel('Average Total Bill ($)', fontsize=14)  
for p in bar_plot.patches:  
    bar_plot.annotate(  
        f'{p.get_height():.2f}',                         
        (p.get_x() + p.get_width() / 2., p.get_height()),  
        ha='center', va='bottom', fontsize=10           
    )  
plt.show()
```

```python
import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D  
import seaborn as sns  
import pandas as pd  

# Load the iris dataset  
iris = sns.load_dataset("iris")  

# Create a 3D scatter plot  
fig = plt.figure(figsize=(10, 7))  
ax = fig.add_subplot(111, projection='3d')  

# Scatter points  
scatter = ax.scatter(iris['sepal_length'], iris['sepal_width'], iris['petal_length'],   
                     c=iris['species'].astype('category').cat.codes,   
                     cmap='viridis',   
                     s=100)  # Size of points  

# Add color bar which maps the species to colors  
legend1 = ax.legend(*scatter.legend_elements(), title="Species")  
ax.add_artist(legend1)  

# Customize the plot  
ax.set_title('3D Scatter Plot of Iris Dataset')  
ax.set_xlabel('Sepal Length')  
ax.set_ylabel('Sepal Width')  
ax.set_zlabel('Petal Length')  

# Show the plot  
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose  
import pandas as pd  
import matplotlib.pyplot as plt  

# Create a time series dataset  
date_rng = pd.date_range(start='2020-01-01', end='2020-12-31', freq='D')  
data = pd.DataFrame(date_rng, columns=['date'])  
data['data'] = pd.Series(range(1, len(data) + 1))  

# Set date as index  
data.set_index('date', inplace=True)  

# Decompose the time series  
decomposition = seasonal_decompose(data['data'], model='additive')  
fig = decomposition.plot()  
plt.show()
```

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Data for the radar plot  
categories = ['A', 'B', 'C', 'D']  
values = [4, 3, 2, 5]  

# Number of variables  
N = len(categories)  

# Compute angle for each axis  
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()  

# The plot is a circular graph, so we need to "complete the loop"  
values += values[:1]  
angles += angles[:1]  

# Create the radar plot  
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))  
ax.fill(angles, values, color='blue', alpha=0.25)  
ax.set_yticklabels([])  
ax.set_xticks(angles[:-1])  
ax.set_xticklabels(categories)  
plt.title("Radar Plot Example")  
plt.show()
```

```python
import plotly.express as px  
import pandas as pd   
data = {  
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Clothing',   
                 'Clothing', 'Clothing', 'Home & Kitchen', 'Home & Kitchen',   
                 'Books', 'Books', 'Toys', 'Toys'],  
    'Subcategory': ['Smartphones', 'Laptops', 'Tablets',   
                    'Men\'s Wear', 'Women\'s Wear', 'Accessories',   
                    'Kitchen Appliances', 'Furniture',   
                    'Fiction', 'Non-Fiction',   
                    'Educational Toys', 'Outdoor Toys'],  
    'Sales': [150, 100, 50,   
              80, 120, 30,   
              90, 60,   
              70, 30,   
              40, 20]  
}  
df = pd.DataFrame(data)   
fig = px.treemap(df,   
                 path=['Category', 'Subcategory'],  # Define categories structure  
                 values='Sales',                     # Values for rectangle size  
                 title='Treemap of Retail Sales by Product Categories',  
                 color='Sales',                      # Color based on sales values  
                 color_continuous_scale='Blues')    # Color scale  
fig.show()
```
